```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.linear_model
import sklearn.metrics
np.array([i%2 for i in range(10000)]).var()
```

```python
data_train = pd.read_csv('kc_house_train_data.csv')
data_test = pd.read_csv('kc_house_test_data.csv')
data_train.head()
```

```python
data_train['bedrooms_squared'] = data_train['bedrooms']**2
data_train['bed_bath_rooms'] = data_train['bedrooms']*data_train['bathrooms']
data_train['log_sqft_living'] = np.log(data_train['sqft_living'])
data_train['lat_plus_long'] = data_train['lat'] + data_train['long']

data_test['bedrooms_squared'] = data_test['bedrooms']**2
data_test['bed_bath_rooms'] = data_test['bedrooms']*data_test['bathrooms']
data_test['log_sqft_living'] = np.log(data_test['sqft_living'])
data_test['lat_plus_long'] = data_test['lat'] + data_train['long']
data_test.head()
```

```python
print(data_test['bedrooms_squared'].mean())
print(data_test['bed_bath_rooms'].mean())
print(data_test['log_sqft_living'].mean())
print(data_test['lat_plus_long'].mean())
```

```python
feature1 = ['sqft_living','bedrooms','bathrooms','lat','long']
feature2 = ['sqft_living','bedrooms','bathrooms','lat','long','bed_bath_rooms']
feature3 = ['sqft_living','bedrooms','bathrooms','lat','long','bed_bath_rooms','bedrooms_squared','log_sqft_living','lat_plus_long']
```

```python
x1_train = data_train[feature1]
x1_test = data_test[feature1]
x2_train = data_train[feature2]
x2_test = data_test[feature2]
x3_train = data_train[feature3]
x3_test = data_test[feature3]
y_train = data_train['price']
y_test = data_test['price']
```

```python
clf = sklearn.linear_model.LinearRegression()
clf.fit(x1_train, y_train)
print( clf.coef_ )
print( clf.intercept_ )
print('train:', sklearn.metrics.mean_squared_error(y_train, clf.predict(x1_train)) * len(y_test) )
print('test:', sklearn.metrics.mean_squared_error(y_test, clf.predict(x1_test)) * len(y_test) )
```

```python
clf = sklearn.linear_model.LinearRegression()
clf.fit(x2_train, y_train)
print( clf.coef_ )
print( clf.intercept_ )
print('train:', sklearn.metrics.mean_squared_error(y_train, clf.predict(x2_train)) * len(y_test) )
print('test:', sklearn.metrics.mean_squared_error(y_test, clf.predict(x2_test)) * len(y_test) )
```

```python
clf = sklearn.linear_model.LinearRegression()
clf.fit(x3_train, y_train)
print( clf.coef_ )
print( clf.intercept_ )
print('train:', sklearn.metrics.mean_squared_error(y_train, clf.predict(x3_train)) * len(y_test) )
print('test:', sklearn.metrics.mean_squared_error(y_test, clf.predict(x3_test)) * len(y_test) )
```
